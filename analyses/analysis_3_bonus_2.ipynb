```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ring_attractor import RingAttractor
```

```python
ring = RingAttractor(n=256, noise=np.linspace(0, 2.5e-3, 25)[11], fixed_points_number=16, plot=True, time=10000, random_seed=1942)
ring.simulate()
```

```python
# Divergence between 'just before the bump' and 'just into the bump'
print(ring.kl_divergence(start_1=2000, start_2=5000, lenght=2000)) # Normal divergence
print(ring.kl_divergence_1(start_1=2000, start_2=5000, lenght=2000)) # Divergence with scale fixed to 1
print(ring.kl_divergence_mean(start_1=2000, start_2=5000, lenght=2000)) # Divergence taking means
```

```python
# Divergence between 'just before the bump' and 'been in the bump for a while'
print(ring.kl_divergence(2000, 8000, 2000))
print(ring.kl_divergence_1(2000, 8000, 2000))
print(ring.kl_divergence_mean(2000, 8000, 2000))
```

```python
# Divergence between 'just into the bump' and 'been in the bump for a while'
print(ring.kl_divergence(5000, 8000, 2000))
print(ring.kl_divergence_1(5000, 8000, 2000))
print(ring.kl_divergence_mean(5000, 8000, 2000))
```

The difference between the first two and the last case should be bigger, but probably we're ok with the first way of computing the divergence.
